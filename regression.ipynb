```python
# load and prepare data
import pandas as pd

df = pd.read_csv("data.csv")
df = df.dropna()
df.head()
```

```python
# define variables
import statsmodels.api as sm

X = df[["Rating", "Events Played", "Ratings x Events"]]
X = sm.add_constant(X)

y = df["Points"]
```

```python
# run regression
model = sm.OLS(y, X).fit()

print(model.summary())
```

```python
# export as json
import json

model_json = {
    "intercept": float(model.params["const"]),
    "coefficients": {
        "Rating": float(model.params["Rating"]),
        "Events_Played": float(model.params["Events Played"]),
        "Rating_x_Events": float(model.params["Ratings x Events"])
    }
}

with open("regression_model.json", "w") as f:
    json.dump(model_json, f, indent=4)
```
